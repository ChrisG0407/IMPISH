```python
import csv
import matplotlib.pyplot as plt
import numpy as np

RRR = 'csv_data_folders/No_Reflector_RRR_GAGG_SipmOut.csv'
RRP = 'csv_data_folders/No_Reflector_RRP_GAGG_SipmOut.csv'
PPP = 'csv_data_folders/No_Reflector_PPP_GAGG_SipmOut.csv'
PPR = 'csv_data_folders/No_Reflector_PPR_GAGG_SipmOut.csv'

datRRR = list()
datRRP = list()
datPPP = list()
datPPR = list()

def read_dat_file(filename):
    with open(filename, newline='') as csvfile:
        nums = csv.reader(csvfile, delimiter=' ')
        count = 0
        dat = list()
        for num in nums:
            if count%11!=0:
                dat.append(int(num[0]))
            else:
                dat.append(num[0])
            count += 1
    return dat

datRRR = read_dat_file(RRR)
datRRP = read_dat_file(RRP)
datPPP = read_dat_file(PPP)
datPPR = read_dat_file(PPR)
```

```python
dx = [40,34,38,22,16,10,4] #for first 7
dz = [40,35,30,25,20,15,10,5] #for second 8
dx2 = [4,10,16,22,28,34,40] #for third 7
dz2 = [5,10,15,20,25,30,35] #for last 7

lColEff_RRR = list()
lColEff_RRP = list()
lColEff_PPP = list()
lColEff_PPR = list()

def calc_eff(data):
    lColEff = list()
    for i in range(0,len(data),11):
        sipm_sum = 0
        scint_sum = 0
        for j in range(i+1,i+5):
            sipm_sum += data[j]
            scint_sum += data[j+5]
        lColEff.append(sipm_sum/scint_sum)
    return lColEff

lColEff_RRR = calc_eff(datRRR)
lColEff_RRP = calc_eff(datRRP)
lColEff_PPP = calc_eff(datPPP)
lColEff_PPR = calc_eff(datPPR)

print(lColEff_RRR)
print(lColEff_RRP)
print(lColEff_PPP)
print(lColEff_PPR)
```

```python
plt.plot(dx,lColEff_RRR[:7])
plt.plot(dx,lColEff_RRP[:7])
plt.plot(dx,lColEff_PPP[:7])
plt.plot(dx,lColEff_PPR[:7])
plt.legend(['RRR','RRP','PPP','PPR'])
plt.title('Light Collection Efficiency W/ Changing X Dimension & Constant 5x40mm2 yz Faces (Small Face Rod to Plate)')
plt.xlabel('x length (mm)')
plt.ylabel('Light Collection Efficiency')
plt.show()
```
